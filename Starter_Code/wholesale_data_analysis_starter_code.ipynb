## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
raw_data_df['line_subtotal'] = raw_data_df['unit_price'] * raw_data_df['qty']

# Print 10 row header
print(raw_data_df.head(10)[['unit_price', 'qty', 'line_subtotal']])
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

# Create total weight column.
raw_data_df['total_weight'] = raw_data_df['unit_weight'] * raw_data_df['qty']

# Calculate shipping price based on total weight.
# Create the 'shipping_price' column
raw_data_df['shipping_price'] = raw_data_df['total_weight'].apply(lambda x: 7 * x if x > 50 else 10 * x)

# Print partial df
print(raw_data_df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(10))
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
raw_data_df['line_price'] = (raw_data_df['line_subtotal'] + raw_data_df['shipping_price']) * 1.0925

# Round total_price to 2 decimal places
raw_data_df['line_price'] = raw_data_df['line_price'].round(2)

# Print partial df
print(raw_data_df[['line_subtotal', 'shipping_price', 'line_price']].head(10))
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
raw_data_df['line_cost'] = (raw_data_df['unit_cost'] * raw_data_df['qty']) + raw_data_df['shipping_price']

# Round line_cost to 2 decimal places
raw_data_df['line_cost'] = raw_data_df['line_cost'].round(2)

# Display the relevant columns to verify
raw_data_df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
raw_data_df['line_profit'] = raw_data_df['line_price'] - raw_data_df['line_cost']
raw_data_df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# Step 1: Group by client_id and count the number of entries for each client
top_clients_by_entries = raw_data_df['client_id'].value_counts().head(5).reset_index()

# Step 2: Rename columns for clarity
top_clients_by_entries.columns = ['Client ID', 'Number of Entries']

# Step 1: Filter the raw_data_df for the clients in top_clients_by_entries
filtered_df = raw_data_df[raw_data_df['client_id'].isin(top_clients_by_entries['Client ID'])]

# Step 2: Calculate total spending for each top client
client_spending = filtered_df.groupby('client_id')['line_price'].sum().reset_index()

# Step 3: Rename columns for clarity
client_spending.columns = ['Client ID', 'Total Spending']

# Step 4: Print the results
print(client_spending)

```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

# Filter the raw_data_df for the top 5 clients
filtered_df = raw_data_df[raw_data_df['client_id'].isin(top_clients_by_entries['Client ID'])]

# Group by client_id and calculate the required totals
client_summary = filtered_df.groupby('client_id').agg({
    'qty': 'sum',                # Total units purchased
    'shipping_price': 'sum',     # Total shipping price
    'line_price': 'sum',         # Total revenue
    'line_profit': 'sum'         # Total profit
}).reset_index()


# Print the header of the summary DataFrame
client_summary.head()
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Rename the columns for clarity
client_summary.columns = [
    'Client ID',
    'Units',
    'Shipping (millions)',
    'Total Revenue (millions)',
    'Total Profit (millions)'
]
# Define the money columns. 
money_columns = ['Shipping (millions)', 'Total Revenue (millions)', 'Total Profit (millions)']

#Convert objects to int64.
for col in money_columns:
    client_summary[col] = pd.to_numeric(client_summary[col], errors='coerce')

# List of columns to convert
columns_to_convert = ['Shipping (millions)', 'Total Revenue (millions)', 'Total Profit (millions)']


# Define a function that converts a dollar amount to millions.
def convert_to_millions(value):
    return value / 1000000


# Apply the currency_format_millions function to only the money columns. 
for col in money_columns:
    client_summary[col] = client_summary[col].apply(convert_to_millions)


# Rename the columns to reflect the change in the money format. 
#completed in line 3

# Display the formatted data.
client_summary.head()


```
